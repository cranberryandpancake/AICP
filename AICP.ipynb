### 06 | 인공지능을 위한 조건문

#### 1. if문 살펴보기

```python
num = 15
if num > 10:
  print("10보다 큽니다.")
```

#### 2. if else문 살펴보기

```python
num2 = 10
if num2 % 2 == 0:
  print("짝수입니다.")
else:
  print("홀수입니다.")
```

#### 3. 홀짝 구별하는 함수 만들기

```python
def number_check(num):
  if num % 2 == 0:
    print("짝수입니다.")
  else:
    print("홀수입니다.")
```

```python
number_check(10)
```

```python
number_check(11)
```

```python
def add_func(a, b):
  result = a + b
  return result
```

```python
num_a = 10
num_b = 15
minmax = add_func(num_a, num_b)
minmax
```

```python
def add_sub(a, b):
  result_add = a + b
  result_sub = a - b
  return result_add, result_sub
```

```python
a = 10
b = 20
c = add_sub(a, b)
c
```

```python
d, e = add_sub(100, 200)
print(d)
print(e)
```
